### Bubble Sort

#### Descripción:
Bubble Sort es un algoritmo simple que recorre repetidamente la lista, compara elementos adyacentes y los intercambia si están en el orden incorrecto. Este proceso se repite hasta que la lista esté ordenada.

#### Algoritmo:
1. Recorre la lista comparando elementos adyacentes.
2. Si están en el orden incorrecto, los intercambia.
3. Repite el proceso hasta que la lista esté ordenada.

**Ejemplo**:
Para la lista `[3, 1, 9]`:

- Primera pasada:
  - Compara 3 y 1: están en el orden incorrecto, se intercambian → `[1, 3, 9]`.
  - Compara 3 y 9: están en el orden correcto, no se intercambian.
- Segunda pasada:
  - Compara 1 y 3: están en el orden correcto, no se intercambian.
  - Compara 3 y 9: están en el orden correcto, no se intercambian.
- La lista ya está ordenada: `[1, 3, 9]`.

**Resultado Final**: `[1, 3, 9]`.

#### Complejidad:
- Tiempo: 
  - Peor caso: O(n^2) (cuando la lista está en orden inverso).
  - Mejor caso: O(n) (cuando la lista ya está ordenada).
- Espacio: O(1) (ordenamiento in-place).



### Merge Sort

#### Descripción:
Merge Sort es un algoritmo de ordenamiento divide y vencerás. Divide la lista en dos mitades, ordena cada mitad recursivamente y luego combina las dos mitades ordenadas.

#### Algoritmo:
1. Divide la lista en dos mitades.
2. Ordena cada mitad recursivamente.
3. Combina las dos mitades ordenadas.

**Ejemplo**:
Para la lista `[1, 9, 8, 2, 3, 4, 5, 7, 6]`:

- División:
  - Primera mitad: `[1, 9, 8, 2]`.
  - Segunda mitad: `[3, 4, 5, 7, 6]`.
- Ordenación recursiva:
  - Primera mitad se divide y ordena: `[1, 2, 8, 9]`.
  - Segunda mitad se divide y ordena: `[3, 4, 5, 6, 7]`.
- Combinación:
  - Combina `[1, 2, 8, 9]` y `[3, 4, 5, 6, 7]` en `[1, 2, 3, 4, 5, 6, 7, 8, 9]`.

**Resultado Final**: `[1, 2, 3, 4, 5, 6, 7, 8, 9]`.

#### Complejidad:
- Tiempo: O(n log n) en todos los casos.
- Espacio: O(n) (se requiere espacio adicional para combinar las mitades).

### Quick Sort

#### Descripción
Quick Sort es otro algoritmo divide y vencerás. Selecciona un elemento como pivote y particiona la lista en dos sublistas: una con elementos menores que el pivote y otra con elementos mayores. Luego, ordena recursivamente las sublistas.

#### Algoritmo
1. Elige un pivote (por ejemplo, el último elemento).
2. Particiona la lista en dos: elementos menores que el pivote y elementos mayores que el pivote.
3. Ordena recursivamente las dos sublistas.

**Ejemplo**:
Para la lista `[8, 2, 4]`:

- Pivote 4.
- Partición:
  - Elementos menores que 4: `[2]`.
  - Elementos mayores que 4: `[8]`.
- Ordenación recursiva:
  - Sublista `[2]` ya está ordenada.
  - Sublista `[8]` ya está ordenada.
- Combinación:
  - Combina `[2]`, `[4]` y `[8]` en `[2, 4, 8]`.

**Resultado Final**: `[2, 4, 8]`.

#### Complejidad
- Tiempo 
  - Promedio: O(n log n).
  - Peor caso: O(n^2) (raro, depende de la elección del pivote).
- Espacio: O(log n) (debido a la recursión).

### Insertion Sort

#### Descripción
Insertion Sort construye la lista ordenada uno por uno, insertando cada elemento en su posición correcta en la parte ya ordenada de la lista.

#### Algoritmo
1. Recorre la lista desde el segundo elemento.
2. Inserta cada elemento en su posición correcta en la parte ya ordenada de la lista.

**Ejemplo**:
Para la lista `[10, 90, 80, 20, 30, 40, 50, 70, 60]`:

- Primera pasada
  - Inserta 90 en su posición correcta: `[10, 90, 80, 20, 30, 40, 50, 70, 60]`.
- Segunda pasada
  - Inserta 80 en su posición correcta: `[10, 80, 90, 20, 30, 40, 50, 70, 60]`.
- Tercera pasada
  - Inserta 20 en su posición correcta: `[10, 20, 80, 90, 30, 40, 50, 70, 60]`.
- Cuarta pasada
  - Inserta 30 en su posición correcta: `[10, 20, 30, 80, 90, 40, 50, 70, 60]`.
- Quinta pasada
  - Inserta 40 en su posición correcta: `[10, 20, 30, 40, 80, 90, 50, 70, 60]`.
- Sexta pasada
  - Inserta 50 en su posición correcta: `[10, 20, 30, 40, 50, 80, 90, 70, 60]`.
- Séptima pasada
  - Inserta 70 en su posición correcta: `[10, 20, 30, 40, 50, 70, 80, 90, 60]`.
- Octava pasada
  - Inserta 60 en su posición correcta: `[10, 20, 30, 40, 50, 60, 70, 80, 90]`.

**Resultado Final**: `[10, 20, 30, 40, 50, 60, 70, 80, 90]`.

#### Complejidad
- Tiempo 
  - Peor caso: O(n^2) (cuando la lista está en orden inverso).
  - Mejor caso: O(n) (cuando la lista ya está ordenada).
- Espacio: O(1) (ordenamiento in-place).

### Selection Sort

Descripción
Selection Sort divide la lista en dos partes: una parte ordenada y una parte desordenada. En cada iteración, encuentra el elemento más pequeño en la parte desordenada y lo intercambia con el primer elemento de la parte desordenada.

Algoritmo
1. Encuentra el elemento más pequeño en la lista.
2. Intercámbialo con el primer elemento.
3. Repite el proceso para el resto de la lista.

**Ejemplo**:
Para la lista `[11, 91, 81, 21, 31, 41, 51, 71, 61]`:

- Primera pasada
  - El elemento más pequeño es 11, ya está en la posición correcta.
- Segunda pasada
  - El elemento más pequeño es 21, se intercambia con 91: `[11, 21, 81, 91, 31, 41, 51, 71, 61]`.
- Tercera pasada
  - El elemento más pequeño es 31, se intercambia con 81: `[11, 21, 31, 91, 81, 41, 51, 71, 61]`.
- Cuarta pasada
  - El elemento más pequeño es 41, se intercambia con 91: `[11, 21, 31, 41, 81, 91, 51, 71, 61]`.
- Quinta pasada
  - El elemento más pequeño es 51, se intercambia con 81: `[11, 21, 31, 41, 51, 91, 81, 71, 61]`.
- Sexta pasada
  - El elemento más pequeño es 61, se intercambia con 91: `[11, 21, 31, 41, 51, 61, 81, 71, 91]`.
- Séptima pasada
  - El elemento más pequeño es 71, se intercambia con 81: `[11, 21, 31, 41, 51, 61, 71, 81, 91]`.
- Octava pasada
  - El elemento más pequeño es 81, ya está en la posición correcta.

**Resultado Final**: `[11, 21, 31, 41, 51, 61, 71, 81, 91]`.

Complejidad
- Tiempo: O(n^2) en todos los casos.
- Espacio: O(1) (ordenamiento in-place).

## Cuándo usar cada algoritmo de ordenamiento

### Bubble Sort y Selection Sort  
No recomendados en la mayoría de los casos, ya que su complejidad O(n²) los hace ineficientes en listas grandes.  
Casos de uso posibles:  
- Educación y enseñanza: Son útiles para aprender los fundamentos del ordenamiento, ya que son fáciles de entender e implementar.  
- Pequeñas listas donde la eficiencia no es un problema.  

### Insertion Sort  
Ideal cuando los datos están casi ordenados. Su mejor caso es O(n), lo que lo hace rápido en listas semiordenadas.  
No recomendado en listas grandes y desordenadas, ya que su peor caso es O(n²).  
Casos de uso:  
- Pequeñas listas con pocos elementos desordenados.  
- Cuando se requiere estabilidad (mantener el orden relativo de elementos iguales).  
- Aplicaciones en tiempo real que necesitan ordenamientos rápidos en datos que llegan progresivamente.  

### Merge Sort  
Eficiente en todos los casos con O(n log n) de complejidad, pero requiere memoria adicional.  
No es la mejor opción si la memoria es un problema.  
Casos de uso:  
- Cuando se necesita estabilidad (mantener el orden relativo de elementos iguales).  
- Cuando se trabaja con estructuras de datos enlazadas.  
- Procesamiento de grandes volúmenes de datos donde se puede dividir y conquistar.  

### Quick Sort  
Rápido en la mayoría de los casos con O(n log n) en promedio.  
Peor caso O(n²) si se elige mal el pivote, aunque puede optimizarse con técnicas como Random Pivoting o Median of Three.  
Casos de uso:  
- Cuando se necesita velocidad y no se requiere estabilidad.  
- Aplicaciones generales en listas grandes donde la memoria es limitada.  
- Ordenamiento en bases de datos o procesamiento de grandes volúmenes de información.  

### Resumen: ¿Cuál usar según la situación?
| Situación | Algoritmo recomendado |
|-----------|----------------------|
| Lista pequeña y sin restricciones | Cualquiera, pero Quick Sort o Merge Sort son mejores. |
| Lista ya ordenada o casi ordenada | Insertion Sort (O(n)). |
| Lista desordenada de gran tamaño | Quick Sort o Merge Sort. |
| Necesidad de estabilidad (mantener orden de elementos iguales) | Merge Sort o Insertion Sort. |
| Memoria limitada | Quick Sort (ya que Merge Sort requiere memoria extra). |
| Enseñanza o propósitos educativos | Bubble Sort o Selection Sort. |
